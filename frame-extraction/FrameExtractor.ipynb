```python
dir_path = "D:\\RESEARCH\\Dataset\\Database"

P_dir = os.path.join(dir_path,"vPorn")
NP_dir = os.path.join(dir_path,"vNonPornEasy")
NPD_dir = os.path.join(dir_path,"vNonPornDifficulty")
P_list = os.listdir(P_dir)
NP_list = os.listdir(NP_dir)
NPD_list = os.listdir(NPD_dir)
Paddress = []
NPaddress = []
NPDaddress = []
for i in P_list:
    Paddress.append(os.path.join(P_dir ,i))
for i in NP_list:
    NPaddress.append(os.path.join(NP_dir ,i))
for i in NPD_list:
    NPDaddress.append(os.path.join(NPD_dir ,i))      
Psave_path = "D:\\RESEARCH\\Dataset\\ExtractedFrames\\PFrames\\"
NPsave_path = "D:\\RESEARCH\\Dataset\\ExtractedFrames\\NPFrames\\"
NPDsave_path = "D:\\RESEARCH\\Dataset\\ExtractedFrames\\NPDFrames\\"
#print(NPDaddress)
```

```python
Paddress =sorted(Paddress, key=lambda x: int(x.split('\\')[-1][5:-4]))
NPaddress = sorted(NPaddress, key=lambda y: int(y.split('\\')[-1][8:-4]))
NPDaddress = sorted(NPDaddress, key=lambda x: int(x.split('\\')[-1][8:-4]))
```

```python
for video_index ,add in enumerate(Paddress[255:]):
    video_index = video_index+255
    y = add.split("\\")[-1]
    name = y
    name = name[:-4]
    Video = cv2.VideoCapture(add)
    print("Accessing Video : ", add,"\n")
    print("Video dimension : ({},{},3)".format(Video.get(3),Video.get(4)))
    frame_rate = int(Video.get(5))
    print("Frame Rate : {}".format(frame_rate))
    if frame_rate<=4:
        continue
    frames = int(frame_rate/4)
    print("Extraction of FRAMES AT : {} ".format(frames))
    total = 1
    sec_frame_count = 1
    total_frame_count=0
    try:
        os.mkdir(Psave_path+"frames_"+name)
    except:
        None    
    while Video.isOpened():
        ret,frame = Video.read()
        if ret != True:
            break
        resized_frame = cv2.resize(frame,(224,224))
        sec_frame_count+=1
        if sec_frame_count>frame_rate:
            sec_frame_count = 1  
            total += 1
        if int(sec_frame_count%frames) == 0:
            total_frame_count +=1
            #print("Total_Frame_Count: {}".format(total_frame_count))
            #cv2_imshow(resized_frame)
            cv2.imwrite(Psave_path+"frames_"+name+"\\"+name+"_{}_{}_{}.jpg".format(total_frame_count,total,sec_frame_count//frames),resized_frame)
            #print("Frame {} >> saved successfully.".format(sec_frame_count/frames))
    print("Video {}-->{} Processed".format(video_index,add))
    Video.release()
    cv2.destroyAllWindows()
```
